## Bab 9. Algoritma Searching

```python
#Linear Search

def linear_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1
    
data = [12, 5, 7, 20, 9, 15]
target = 20
hasil = linear_search(data, target)

if hasil != -1:
    print("Data ditemukan pada indeks", hasil)
else:
    print("Data tidak ditemukan")
```

```python
#Binary Search

def binary_search(data, target):
    kiri = 0
    kanan = len(data) - 1
    
    while kiri <= kanan:
        tengah = (kiri + kanan) // 2
        if data[tengah] == target:
            return tengah
        elif data[tengah] < target:
            kiri = tengah + 1
        else:
            kanan = tengah - 1
        return -1
        
data = [2, 4, 6, 8, 10, 12, 14]
target = 10
print(binary_search(data, target))
```

 ## Bab 10. Algoritma Sorting

```python
#Bubble Sort

def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
        return data
        
angka = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(angka))
```

```python
#Selection Sort

def selection_sort(data):
    n = len(data)
    for i in range(n):
        indeks_min = i
        for j in range(i + 1, n):
            if data[j] < data[indeks_min]:
                indeks_min = j
        data[i], data[indeks_min] = data[indeks_min], data[i]
    return data
    
angka = [29, 10, 14, 37, 13]
print(selection_sort(angka))
```

```python
#Insertion Sort

def insertion_sort(data):
    for i in range(1, len(data)):
        kunci = data[i]
        j = i - 1
        while j >= 0 and data[j] > kunci:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = kunci
    return data
    
angka = [12, 11, 13, 5, 6]
print(insertion_sort(angka))
```

```python
#Merge Sort

def merge_sort(data):
    if len(data) <= 1:
        return data
        
    tengah = len(data) // 2
    kiri = merge_sort(data[:tengah])
    kanan = merge_sort(data[tengah:])
    
    return merge(kiri, kanan)
    
def merge(kiri, kanan):
    hasil = []
    i = j = 0
    
    while i < len(kiri) and j < len(kanan):
        if kiri[i] < kanan[j]:
            hasil.append(kiri[i])
            i += 1
        else:
            hasil.append(kanan[j])
            j += 1
            
    hasil.extend(kiri[i:])
    hasil.extend(kanan[j:])
    return hasil
    
angka = [38, 27, 43, 3, 9, 82, 10]
print(merge_sort(angka))
```

```python
#Quick Sort

def quick_sort(data):
    if len(data) <= 1:
        return data
        
    pivot = data[len(data) // 2]
    kiri = [x for x in data if x < pivot]
    tengah = [x for x in data if x == pivot]
    kanan = [x for x in data if x > pivot]
    
    return quick_sort(kiri) + tengah + quick_sort(kanan)
    
angka = [10, 7, 8, 9, 1, 5]
print(quick_sort(angka))
```

## Bab 13. Divide and Conquer

```python
# Divide and Conquer adalah strategi penyelesaian masalah dengan membagi masalah besar menjadi submasalah yang lebih kecil, menyelesaikannya, lalu menggabungkan hasilnya.
```

```python
#Contoh: Mencari Nilai Maksimum

def cari_maksimum(data, kiri, kanan):
    if kiri == kanan:
        return data[kiri]
        
    tengah = (kiri + kanan) // 2
    maksimum_kiri = cari_maksimum(data, kiri, tengah)
    maksimum_kanan = cari_maksimum(data, tengah + 1, kanan)

    return max(maksimum_kiri, maksimum_kanan)
    
data = [7, 2, 9, 4, 11, 6]
print(cari_maksimum(data, 0, len(data) - 1))
```

## Bab 14. Greedy Algorithm

```python
# Greedy Algorithm memilih solusi terbaik pada setiap langkah lokal dengan harapan memperoleh solusi global yang baik.
```

```python
#Contoh: Penukaran Uang

def greedy_coin_change(jumlah, koin):
    hasil = []
    koin.sort(reverse=True)
    
    for nilai in koin:
        while jumlah >= nilai:
            jumlah -= nilai
            hasil.append(nilai)
            
    return hasil
    
koin = [1000, 500, 200, 100]
jumlah = 2700
print(greedy_coin_change(jumlah, koin))
```

## Bab 15. Dynamic Programming

```python
def fibonacci_dp(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_dp(n - 1, memo) + fibonacci_dp(n - 2, memo)
    return memo[n]
    
print(fibonacci_dp(40))
```

```python
#Contoh: Knapsack Sederhana

def knapsack(berat, nilai, kapasitas):
    n = len(nilai)
    dp = [[0 for _ in range(kapasitas + 1)] for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, kapasitas + 1):
            if berat[i - 1] <= w:
                dp[i][w] = max(nilai[i - 1] + dp[i - 1][w - berat[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    return dp[n][kapasitas]
    
berat = [2, 3, 4, 5]
nilai = [3, 4, 5, 6]
kapasitas = 5
print(knapsack(berat, nilai, kapasitas))
```

## Bab 16. Backtracking

```python
# COntoh: N_Queen 4x4

def aman(board, baris, kolom):
    for i in range(baris):
        if board[i] == kolom or abs(board[i] - kolom) == abs(i - baris):
            return False
    return True


def n_queen(n, baris=0, board=[]):
    if baris == n:
        print(board)
        return

    for kolom in range(n):
        if aman(board, baris, kolom):
            n_queen(n, baris + 1, board + [kolom])


n_queen(4)
```

## Bab 17. Graph Algorithms

```python
# Representasi Graph dengan Dictionary

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print(graph)
```

```python
from collections import deque


def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()

        if node not in visited:
            print(node, end=' ')
            visited.add(node)

            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)


bfs(graph, 'A')
```

```python
# Depth First Search

def dfs(graph, node, visited=None):
    if visited is None:
        visited = set()

    if node not in visited:
        print(node, end=' ')
        visited.add(node)

        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)


dfs(graph, 'A')
```

```python
# Dijkstra

import heapq


def dijkstra(graph, start):
    jarak = {node: float('inf') for node in graph}
    jarak[start] = 0

    pq = [(0, start)]

    while pq:
        jarak_sekarang, node = heapq.heappop(pq)

        if jarak_sekarang > jarak[node]:
            continue

        for tetangga, bobot in graph[node].items():
            jarak_baru = jarak_sekarang + bobot

            if jarak_baru < jarak[tetangga]:
                jarak[tetangga] = jarak_baru
                heapq.heappush(pq, (jarak_baru, tetangga))

    return jarak


graph_berbobot = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 1, 'D': 5},
    'C': {'A': 2, 'B': 1, 'D': 8, 'E': 10},
    'D': {'B': 5, 'C': 8, 'E': 2},
    'E': {'C': 10, 'D': 2}
}

print(dijkstra(graph_berbobot, 'A'))
```
